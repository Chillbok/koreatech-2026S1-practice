```python
p, q = True, False
print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")
```

```python
p, q = True, False
left = not(p and q)
right = (not p) or (not q)
print(f"De Morgan's Equivalent: {left == right}")
```

```python
cases = [True, False]
is_tautology = all(p or (not p) for p in cases)
print(f"is p v ~p a Tautology? {is_tautology}")
```

```python
def implies(p, q) :
    return (not p) or q

print(f"T->F: {implies(True, False)}")
print(f"F->T: {implies(False, True)}")
```

```python
def iff(p, q) :
    return p==q

print(f"T iff T: {iff(True, True)}")
print(f"T iff F: {iff(True, False)}")
```

```python
def contrapositive_check(p, q) :
    original = implies(p, q)
    contra = implies(not q, not p)
    return original == contra

print(f"Is Contrapositive same? {contrapositive_check(True, False)}")
```
