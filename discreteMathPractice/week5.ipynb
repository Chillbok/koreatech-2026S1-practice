```python
def P(x):
    return x**2 > x

def Q(n):
    return 8 % n == 0

# Truth Set
domain_z_plus = range(1, 11)
domain_z = range(-10, 11)

truth_set_a = [n for n in domain_z_plus if Q(n)]
truth_set_b = [n for n in domain_z if n != 0 and Q(n)]
print(f"a) Truth set in Z+: {truth_set_a}")
print(f"b) Truth set in Z: {truth_set_b}")

# Universal Quantifier all ∀ (전칭양화사)
D = {1, 2, 3, 4, 5}
universal_check_D = all(P(x) for x in D)
print(f"\n∀x ∈ D, x^2 >= x: {universal_check_D}")

# 반례
R_sample = [2, 1, 0.5, -1]
for x in R_sample:
    if not P(x):
        print(f"Counterexample found: x = {x} (p({x}) is false)")

# Existential Quantifier (∃): any()
E = {5, 6, 7, 8}
existential_check_E = any(m**2 == m for m in E)
print(f"∃m ∈ E such that m^2 = m: {existential_check_E}")

# Universal Conditional Statement ( ∀x, ir P(x) then Q(x) )
# "x > 2 이면, x^2 > 4 이다" 검증
def conditional_check(x):
    if x > 2:
        return x ** 2 > 4

domain_R = [1, 2, 3, 4, 5]
is_universal_conditional_true = all(conditional_check(x) for x in domain_R)
print(f"∀x ∈ R, ( x > 2 ) -> ( x^2 > 4 ): {is_universal_conditional_true}")
```
