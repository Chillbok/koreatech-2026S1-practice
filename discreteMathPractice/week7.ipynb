## 전칭 조건문의 변형

```python
print("n### 3. 전칭 조건문의 변형 실습 (대우, 역, 이) ###")
# 원 명제: ∀x ∈ R, if x > 2 then x^2 > 4

domain_r = [-3, -2, -1, 0, 1, 2, 3, 4, 5]

# 표 헤더 출력
print(f"{'x':>3} | {'P->Q' :>8} | {'~Q->~P(대우)' :>12} | {'Q->P(역)' :>8} | {'~P->~Q(이)' :>10}")
print("-" * 65)

for x in domain_r:
	p = x > 2
	q = x**2 > 4
	
	# 1. Original: P -> Q (not p or q)
	original = (not p) or q
	
	# 2. Contrapositive: ~Q -> ~P (not (not q) or (not q) ===> q or not p)
	Contrapositive = q or (not p)
	
	# 3. Converse: Q -> P (not q or p)
	converse = (not q) or p
	# 4. Inverse: ~P -> ~Q (not(not p) or (not q) ===> p or not q)
	inverse = p or (not q)
	
	print(f"{x:3d} | {str(original):8} | {str(Contrapositive):12} | {str(converse):8} | {str(inverse):10}")
```
