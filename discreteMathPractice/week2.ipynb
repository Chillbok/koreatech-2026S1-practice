```python
def is_function(domain, relation):
    mapping = {}
    for x, y in relation:
        if x in mapping:
            return False, f"실패: 원소 {x}가 두 개 이상의 값 ({mapping[x]}, {y})을 가짐."
        mapping[x] = y

    for x in domain:
        if x not in mapping:
            return False, f"실패: 원소 {x}가 대응하는 값이 없음"
    return True, "성공: 함수가 맞습니다."

A = {2, 4, 6}
R = [(2, 5), (4,1), (6,5)] # 함수
S = [(2,5), (4,1), (4,3)] # 함수 아님
print("R 결과: ", is_function(A, R))
print("S 결과: ", is_function(A, S))
```

세 집합 A, B, C에 대해 $(A \cup B) \cap C$를 코드로 구현해보기

```python
def check_function_properties(domain, codomain, relation):
    starts = [pair[0] for pair in relation]
    is_unique = len(starts) == len(set(starts))
    
    sent_elements = set(starts)
    all_targets_in_B = all(pair[1] in codomain for pair in relation)
    is_existent = (sent_elements == domain) and all_targets_in_B
    return is_existent, is_unique
A_set = {2, 4, 6}
B_set = {1, 3, 5}
R = [(2,5), (4,1), (4,3), (6,5)]

existent, unique = check_function_properties{A_set, B_set, R}
print(f"존재성 만족: {existent}, 유일성 만족: {unique}")
```
