```python
def find_counterexample(limit):
	"""
	명제: 모든 홀수 n에 대해, (n-1)^2은 8로 나누어 떨어진다.
	이 명제가 거짓임을 증명하기 위해 반례를 하나만 찾는다.
	"""
	for n in range(1, limit, 2): # 홀수만 검사
		statement = (n-1) ** 2
		if statement % 8 != 0:
			return f"Counterexample found! n = {n}, (n-1)^2 = {statement} \
		(Not divisible by 8)"
		return "No countexample found in the given range."
print(find_counterexample(20))
```

```python
def check_composite(n):
	"""
	정의: n > 1이고, n = r * s (1 < r, s < n)를 만족하는 r, s가 존재하면 합성수
	"""
	if n <= 1: return "Neither prime nor composite"
	for r in range(2, int(n ** 0.5) + 1):
		if n % r == 0:
			s = n // r
			return f"{n} is Composite because {n} = {r} * {s}"
	return f"{n} is Prime"

print(check_composite(15))
```

```python
def verify_uniqueness(n, d):
	"""
	n = dq + r (0 <= r < d)을 만족하는 쌍이 정말 하나뿐인지 전수 조사
	"""
	result = []
	# 충분한 범위의 q를 조사
	for q in range(n // d - 5, n // d + 5)
		r = n - (d * q);
		if 0 <= r < d:
			result.append((q, r))
	if len(result) == 1:
		return f"Uniqueness Verified! Only one pair exists: {result[0]}"
	else:
		return f"Uniqueness Failed! Multiple pairs found: {result}"

print(verify_uniqueness(19, 4))
```
