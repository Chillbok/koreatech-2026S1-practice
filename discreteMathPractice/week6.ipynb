# 기반 함수들

```python
import math

numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]
triangles_angles = [
    {"id":"T1", "angles" : [60, 60, 60]}, # 합 180
    {"id":"T2", "angles" : [90, 45, 45]}, # 합 180
    {"id":"T3", "angles" : [30, 60, 90]}, # 합 180
]

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0: return False
    return True

def is_odd(n):
    return n % 2 != 0
```

# 1. 전칭 / 존재 명제의 부정 실습

```python
print("### 1. 전칭/존재 명제의 부정 실습 ###")

# a) ∀p ∈ Primes, p is odd (모든 소수는 홀수이다)
primes = [n for n in numbers if is_prime(n)]
all_primes_are_odd = all(is_odd(p) for p in primes)
print(f"모든 소수는 홀수인가? (∀p): {all_primes_are_odd}")

# 부정: ∃p ∈ Primes, p is not odd (홀수가 아닌 소수가 적어도 하나 존재한다)
negation_exists = any(not is_odd(p) for p in primes)
counter_examples = [p for p in primes if not is_odd(p)]

print(f"홀수가 아닌 소수가 존재하는가? (∃p ~Q(p)) : {negation_exists}")
print(f"반례(counter-example): {counter_examples}") # 2가 출력됨
```

# 2. 공허한 참 실습

```python
print("\n### 2. 공허한 참 ( Vacuous Truth) 실습 ###")

# 그릇에 공이 하나도 없는 상황
bowl = []

# 문장: "그릇 안의 모든 공이 파란색이다." (∀x ∈ Bowl, Blue(x))
# P(x) : x는 그릇 안에 있다. (P(x)가 모든 x에 대해 거짓)
is_all_blue = all(ball == "blue" for ball in bowl)
print(f"빈 그릇 안의 공은 모두 파란색인가? (all() 사용): {is_all_blue}")

# 부정: "그릇 안에 파란색이 아닌 공이 적어도 하나 존재한다."
exists_not_blue = any(ball != "blue" for ball in bowl)
print(f"빈 그릇 안에 파란색이 아닌 공이 있는가? (any() 사용): {exists_not_blue}")
print("-> 부정이 거짓이므로 원문장은 '공허하게 참(VAcuously True)' 이다.")
```
