# 1. 정의를 활용한 홀수와 짝수 판별

```python
def check_even_odd(n):
	"""
	정의: n이 정수일 때, n = 2k이면 짝수, n = 2k+1이면 홀수이다.
	나머지 연산자(%)를 사용하여 확인.
	"""
	if n % 2 == 0:
		return f"{n} is Even (2k form)"
	else:
		return f"{n} is Odd (2k + 1 form)"
```

# 2. 유리수 확인

```python
def is_rational(number):
	"""
	정의: r이 유리수이면 r = a/b (a, b는 정수, b != 0)
	Python의 float.as_integer_ratio()는 실수를 정수 바로 반환
	"""
	try:
		a, b = number.as_integer_ratio()
		return True, a, b
	except OverflowError:
		return False, None, None
```

# 3. 가해성(Divisibility) 확인

```python
def is_divisible(n, d):
	"""
	정의: d | n은 n = d * k인 정수 k가 존재함을 의미한다.
	"""
	if d == 0:
		return "Division by zero is undefined."

	if n % d == 0:
		k = n // d
		return f"{d} | {n} is True because {n} = {d} * {k}"
	else:
		return f"{d} | {n} is False"
```

# 4. 나눗셈 알고리즘

```python
def division_algorithm(n, d):
	"""
	정의: n = d * q + r ( 0 <= r < d)을 만족하는 유일한 q, r이 존재한다.
	"""
	if d <= 0:
		return "Divisor d must be a positive integer."

	q = n // d # 몫
	r = n % d # 나머지
	
	return {"dividend": n, "divisor": d, "quotient": q, "remainder": r}
```

## 실습 테스트

```python
if __name__ == "__main__":
	print("=== 1. Even/Odd Check ===")
	print(check_even_odd(7))

	print("=== 2. Rational Number Check ===")
	num = 0.75
	success, a, b = is_rational(num)
	
	if (success):
		print(f"{num} is rational: {a} / {b}")
		
	print("=== 3. Divisibility Check ===")
	print(is_divisible(15, 3))

	print("=== 4. Division Algorithm ===")
	result = division_algorithm(19, 4)
	print(f"19 = {result['divisor']} * {result['quotient']} + {result['remainder']}")
```
